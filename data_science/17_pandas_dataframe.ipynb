```python
class Pattern:
    def __init__(self):
        print('patterns')
    
    def pattern1(n):
        for i in range(n):
            for j in range(n):
                print('*',end='')
            print('\n')

    def pattern2(n):
        for i in range(n+1):
            for j in range(i):
                print('*',end='')
            print('\n')

    def pattern3(n):
        for i in range(n+1):
            for j in range(i):
                print(j+1 ,end='')
            print('\n')

obj = Pattern
obj.pattern3(5)
        
```
