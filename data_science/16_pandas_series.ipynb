### What is Pandas
Pandas is a fast, powerful, flexible and easy to use open source data anaylsis and manipulation tool, built on top of the python programming language.

### Panda Series
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
# Importing Pandas.
import numpy as np
import pandas as pd
```

```python
# series from lists
country =  ['india','Pakithan','USA','Nepal','Srilanka']
pd.Series(country)
```

```python
runs = [13,24,56,78,100]
runs_series = pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,100]
subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
```

```python
# setting a name
marks  = pd.Series(marks,index=subjects,name='Ayush ke marks')
marks
```

```python
marks = {
    'maths'   : 67,
    'english' : 57,
    'science' : 89,
    'hindi'   : 100
}
marks_series = pd.Series(marks,name="ayush")
print(type(marks_series))
marks_series
```
